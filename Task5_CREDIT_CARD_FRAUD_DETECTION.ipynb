# Preparing the data

```python
#Handling the imbalance

df_fraud = df[df['Class']==1]
df_not_fraud = df[df['Class']==0]
df_not_fraud_sampled = df_not_fraud.sample(df_fraud.shape[0],replace= False,random_state=101)
df_balanced = pd.concat([df_not_fraud_sampled,df_fraud],axis=0)
df_balanced = df_balanced.sample(frac=1, replace=False,random_state=101).reset_index(drop=True)

df_balanced
```

```python
#checking the balance of the dataset
fraud_count = df['Class'].value_counts()
fraud_rate = 100*fraud_count/df_balanced.shape[0]
fraud_data = pd.concat([fraud_count , fraud_rate], axis = 1).reset_index()
fraud_data.columns = ["Class",'Count','Percentage']

fraud_data
```

```python
#train test split
from sklearn import *
from sklearn.model_selection import train_test_split

x_train , x_test, y_train, y_split = train_test_split(df_balanced.drop('Class', axis=1),df_balanced['Class'],test_size=0.2, random_state=101)
print(f'''X_train: {x_train.shape})
X_test: {x_test.shape}
y_train: {y_train.shape}
y_test: {y_test.shape}''')
```

# Fitting a Random Model

```python
#logistic model pipe
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier


randomForestModel = Pipeline([
    ('scalar', StandardScaler()),
    ('classifier', RandomForestClassifier())
])

randomForestModel.fit(x_train, y_train)
```

# prediction

```python
#prediction

y_pred_logis = randomForestModel.predict(x_test)
y_pred_logis
```
